```python
metadata = pd.read_csv('../data/AlphaGenome/AlphaGenome_metadata.csv')
metadata
```

```python
metadata['Assay title'].value_counts()
```

```python
metadata['cell_type'] = metadata['biosample_name']
metadata['assay'] = metadata['Assay title']

mask = (metadata['Assay title'] == 'DNase-seq')
metadata.loc[mask, 'assay'] = 'DNase'
mask = (metadata['Assay title'] == 'TF ChIP-seq')
metadata.loc[mask, 'assay'] = metadata.loc[mask, 'transcription_factor']
mask = (metadata['Assay title'] == 'Histone ChIP-seq')
metadata.loc[mask, 'assay'] = metadata.loc[mask, 'histone_mark']

metadata['index'] = metadata.index
```

```python
df_pivot = metadata.pivot_table(
    values="index", 
    index="cell_type", 
    columns="assay", 
    aggfunc=list,
)

df_pivot
```

```python
# assays = ['DNase', 'H3K4me3', 'H3K27ac', 'CTCF']
# df_pivot.loc[df_pivot[assays].notna().all(axis=1), assays]
```

```python
cell_types = ['HepG2', 'HEK293T', 'Jurkat', 'PANC-1', 'HeLa-S3']
cell_names = ['K562', 'HepG2', 'SK-N-SH', 'HCT116', 'A549', 'HEK293', 'Jurkat', 'DND-41', 'Panc1', 'HeLa-S3']
assays = ['DNase', 'H3K4me3', 'H3K27ac', 'CTCF']

df_track = df_pivot.loc[cell_names, assays]
df_track
```

```python
pred_DNase = load_h5('outputs/TFBU_MPRA_18000_AlphaGenome_pred_DNase.h5')
print(pred_DNase.shape)

pred_Histone = load_h5('outputs/TFBU_MPRA_18000_AlphaGenome_pred_Histone.h5')
print(pred_Histone.shape)

pred_TF = load_h5('outputs/TFBU_MPRA_18000_AlphaGenome_pred_TF.h5')
print(pred_TF.shape)
```

```python
MPRA_df = pd.read_csv('../data/TFBU_MPRA/TFBU_MPRA_18000.tsv', sep='\t')
print(MPRA_df.shape)
```

```python
VEF_df = pd.DataFrame()

for i, cell_type in enumerate(cell_names):
    for j, assay in enumerate(assays):
        indice = df_track.loc[cell_type, assay]
        if isinstance(indice, list) and len(indice) > 0:
            if assay == 'DNase':
                indice = [i-713 for i in indice]
                pred_array = pred_DNase
            elif assay in ['H3K4me3', 'H3K27ac']:
                indice = [i-1685 for i in indice]
                pred_array = pred_Histone
            elif assay == 'CTCF':
                indice = [i-2803 for i in indice]
                pred_array = pred_TF
            else:
                VEF_df[f'{cell_type}_{assay}'] = np.nan
                continue
            pred = pred_array[:, indice].mean(1)
            VEF_df[f'{cell_type}_{assay}'] = pred
        else:
            VEF_df[f'{cell_type}_{assay}'] = np.nan

VEF_df.describe()
```

```python
for j, assay in enumerate(assays):
    pearson_df = pd.DataFrame()
    for i, cell_type in enumerate(cell_types):
        for ii, cell_name in enumerate(cell_names):
            pred = VEF_df[f'{cell_name}_{assay}']
            true = MPRA_df[f'{cell_type}']
            r, _ = pearson(pred, true)
            pearson_df.loc[f'{cell_name}_{assay}', f'{cell_type}'] = r
    print(pearson_df)
```

```python
##### log1p

VEF_df = pd.read_csv('../data/TFBU_MPRA/TFBU_MPRA_AG_VEF_raw.tsv', sep='\t')

cell_types = ['HepG2', 'HEK293T', 'Jurkat', 'PANC']
assays = ['DNase', 'H3K4me3', 'H3K27ac', 'CTCF']

cols = [f'{cell_type}_{assay}' for cell_type in cell_types for assay in assays]
x = VEF_df[cols].copy()
x = np.log1p(x)
VEF_df[cols] = x

VEF_df.to_csv('../data/TFBU_MPRA/TFBU_MPRA_AG_VEF_log1p.tsv', sep='\t', index=False)

pearson_df = pd.DataFrame()
for i, cell_type in enumerate(cell_types):
    for j, assay in enumerate(assays):
        pred = VEF_df[f'{cell_type}_{assay}']
        true = MPRA_df[f'{cell_type}']
        r, _ = pearson(pred, true)
        pearson_df.loc[cell_type, assay] = r
print(pearson_df)
```

```python
# æ¯ä¸ª assay å›ºå®š scale factor

VEF_df = pd.read_csv('../data/TFBU_MPRA/TFBU_MPRA_AG_VEF_raw.tsv', sep='\t')

cell_types = ['HepG2', 'HEK293T', 'Jurkat', 'PANC']
assays = ['DNase', 'H3K4me3', 'H3K27ac', 'CTCF']

for assay in assays:
    for cell_type in cell_types:
        if assay == 'DNase':
            scale_factor = 0.1
        else:
            scale_factor = 10
        x = VEF_df[f'{cell_type}_{assay}'].copy()
        x = np.log1p(x / scale_factor)
        VEF_df[f'{cell_type}_{assay}'] = x

VEF_df.to_csv('../data/TFBU_MPRA/TFBU_MPRA_AG_VEF_log1p_scale.tsv', sep='\t', index=False)

pearson_df = pd.DataFrame(dtype=float)
for i, cell_type in enumerate(cell_types):
    for j, assay in enumerate(assays):
        pred = VEF_df[f'{cell_type}_{assay}']
        true = MPRA_df[f'{cell_type}']
        r, _ = pearson(pred, true)
        pearson_df.loc[cell_type, assay] = r
print(pearson_df)
```
