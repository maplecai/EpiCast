```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# ============== é¢„å¤„ç† ==============
cell_types_pred = [f'{i:04b}_pred' for i in range(16)]
plot_df = MPRA_df[cell_types_pred].melt(var_name='cell_type', value_name='activity')
plot_df['cell_type'] = plot_df['cell_type'].str.replace('_pred', '', regex=False)
plot_df.dropna(inplace=True)

for i in range(4):
    plot_df[f'bit_{i}'] = plot_df['cell_type'].str[i]

bit_labels = [
    "Virtual\nDNase",
    "Virtual\nH3K4me3",
    "Virtual\nH3K27ac",
    "Virtual\nCTCF"
]

# ============== è®¡ç®—ç»Ÿè®¡é‡å¹¶å‡†å¤‡ç»˜å›¾æ•°æ® ==============
diffs, pvals, data, positions, colors = [], [], [], [], []

for i, label in enumerate(bit_labels):
    bit_col = f'bit_{i}'
    g0 = plot_df.loc[plot_df[bit_col] == '0', 'activity']
    g1 = plot_df.loc[plot_df[bit_col] == '1', 'activity']

    diffs.append(g1.mean() - g0.mean())
    pvals.append(ttest_ind(g0, g1, equal_var=False).pvalue)

    data += [g0, g1]
    positions += [i*2, i*2+1]
    colors += ['#dddddd', '#555555']

# ============== ç»˜å›¾ ==============
fig, ax = plt.subplots(figsize=(10, 6))

box = ax.boxplot(
    data, positions=positions, widths=0.6, patch_artist=True, showfliers=False,
    boxprops=dict(linewidth=1, edgecolor='black'),
    medianprops=dict(color='black'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
)

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

xtick_labels = [f'{lbl}\n{sign}' for lbl in bit_labels for sign in ['-', '+']]
ax.set_xticks(positions)
ax.set_xticklabels(xtick_labels, fontsize=14)
ax.set_ylabel("Predicted activity", fontsize=14)

# æ·»åŠ  Î” å’Œ p å€¼
for i, (d, p) in enumerate(zip(diffs, pvals)):
    x_mid = i * 2 + 0.5
    y_top = 3
    p_text = f"p < 1.0e-308" if p == 0.0 else f"p = {p:.2e}"
    ax.text(x_mid, y_top, f"Î” = {d:.3f}\n{p_text}", ha='center', va='bottom', fontsize=14)

plt.tight_layout()
plt.savefig('figures/Gosai_MPRA_pseudo_epi_compare_boxplot.png', dpi=400)
plt.show()
```
