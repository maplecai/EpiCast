```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 
x_labels = ['test', 'test+cell type specific']      # 
train_types = ['EpiCast', 'EpiCast(K562+HepG2)', 'EpiCast(K562)', 'CNN-Transformer']

values = np.array([
    [0.816, 0.797, 0.779, 0.813],   # test
    [0.559, 0.592, 0.503, 0.435]    # test + cell type specific
])

# 
sns.set_theme(style="whitegrid", context="talk")
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
plt.subplots_adjust(left=0.2, bottom=0.2, right=0.9, top=0.9)

# x 
x = np.arange(len(x_labels))
num_bars = values.shape[1]
width = 0.2

deep_colors = sns.color_palette("deep")
for i in range(num_bars):
    ax.bar(x + (i - 1.5) * width, values[:, i], width, label=train_types[i], color=deep_colors[i])

# 
ax.set_xlabel('Test Setting')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.set_ylim([0, 1])
ax.set_ylabel('Performance on HCT116')

#  legend 
ax.legend(fontsize=14)
plt.savefig('figures/Gosai_MPRA_EpiCast_Ablation.png', dpi=400)
plt.savefig('figures_pdf/Gosai_MPRA_EpiCast_Ablation.pdf', dpi=400)
plt.show()
```
