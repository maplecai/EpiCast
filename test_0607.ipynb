```python
saved_dir = 'saved/0531_GosaiMPRA_baseline/0531_164123'

y_pred_list = np.load(f'{saved_dir}/y_pred_list.npy')
print(y_pred_list.shape)

main_df = pd.read_csv('/home/hxcai/cell_type_specific_CRE/data/GosaiMPRA/GosaiMPRA_total.csv')
for i, cell_type in enumerate(['HepG2']):
    main_df[f'{cell_type}_pred'] = y_pred_list[:, i]
main_df
```

```python
data_split_dict = {}
data_split_dict['total'] = np.ones(len(main_df), dtype=bool)
data_split_dict['chrom'] = main_df['chr'].isin([f'chr{i}' for i in range(1, 23)])
data_split_dict['train'] = main_df['chr'].isin([f'chr{i}' for i in range(3, 23)])
data_split_dict['valid'] = main_df['chr'].isin(['chr1'])
data_split_dict['test'] = main_df['chr'].isin(['chr2'])
data_split_dict['designed'] = main_df['chr'].isin(['chr0'])


cell_types = ['HepG2', 'K562']
main_df['diff_exp'] = (main_df[cell_types].max(axis=1) - main_df[cell_types].min(axis=1)) > 1
data_split_dict['const'] = ~main_df['diff_exp']
data_split_dict['diff']  = main_df['diff_exp']
```

```python
for cell_type in ['HepG2', 'K562']:
    for cell_type_pred in ['HepG2', 'K562']:
        for data_split in ['total', 'train', 'valid', 'test', 'designed', 'const', 'diff']:
            condition = data_split_dict[data_split]
            r = pearson(main_df[condition][f'{cell_type}'], main_df[condition][f'{cell_type_pred}_pred'])
            print(f'{cell_type} true vs {cell_type_pred} pred, {data_split} data, pearson r = {r:.3f}')
        print()
```

```python
data_split_column = 'chr'
data_split_dict = {
    'train': ['chr'+str(i) for i in range(3,23)],
    'valid': ['chr1'],
    'test': ['chr2'],
    'designed': ['chr0'],
}
# data_split_index_dict = {
#     data_split: main_df[main_df[data_split_column].isin(data_split_dict[data_split])].index
#     for data_split in data_split_dict
# }
data_split_condition_dict = {
    data_split: main_df[data_split_column].isin(data_split_dict[data_split])
    for data_split in data_split_dict
}
```
